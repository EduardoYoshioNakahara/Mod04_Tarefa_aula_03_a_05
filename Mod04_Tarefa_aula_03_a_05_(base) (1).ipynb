## Exercício 1: Vestibular

Considere que a os dados gerados na célula abaixo contêm o número de acertos de 100 alunos em um vestibular para um curso de exatas, divididas pelos respectivos assuntos. Considere que cada assunto possui um número de questões conforme a tabela abaixo:

| assunto | número de questões |
|:---:|:---:|
| Matemática | 24 |
| Português | 18 |
| Geografia | 8 |
| Inglês | 8 |
| História | 8 |
| Física | 12 |
| Química | 12 |

Usando os comandos de operações com DataFrames que você aprendeu na Aula 03, calcule:

1. (operações com escalar) Calcule o percentual de acerto dos alunos por assunto.  
2. (operações entre *DataFrames) Calcule o total de acertos de cada aluno.  
3. Calcule o porcentual geral de cada aluno.  
4. Suponha que a nota de corte para a segunda fase seja 45. Quantos alunos tiveram nota maior que 45?  

````python
import pandas as pd
import numpy as np

np.random.seed(42)
df_mat = pd.DataFrame(np.random.randint(24, size=(100, 1)), columns=['Mat Qt_acertos'])

df_por = pd.DataFrame(np.random.randint(18, size=(100, 1)), columns=['Por Qt_acertos'])

df_geo = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Geo Qt_acertos'])

df_ing = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['Ing Qt_acertos'])

df_his = pd.DataFrame(np.random.randint(8, size=(100, 1)), columns=['His Qt_acertos'])

df_fis = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Fis Qt_acertos'])

df_qui = pd.DataFrame(np.random.randint(12, size=(100, 1)), columns=['Qui Qt_acertos'])
````

````python
# 1) Seu código aqui
tabela_geral = pd.concat([df_mat, df_por, df_geo, df_ing, df_his, df_fis, df_qui], axis=1)


tabela_porcentagem = pd.DataFrame()
tabela_porcentagem['Mat'] = (df_mat/24)*100
tabela_porcentagem['Por'] = (df_por/18)*100
tabela_porcentagem['Geo'] = (df_geo/8)*100
tabela_porcentagem['Ing'] = (df_ing/8)*100
tabela_porcentagem['His'] = (df_his/8)*100
tabela_porcentagem['Fis'] = (df_fis/12)*100
tabela_porcentagem['Qui'] = (df_qui/12)*100

tabela_porcentagem
````

````python
# 2) Seu código aqui
tabela_geral['Acertos'] = (tabela_geral['Mat Qt_acertos'] +
                           tabela_geral['Por Qt_acertos'] +
                           tabela_geral['Geo Qt_acertos'] +
                           tabela_geral['Ing Qt_acertos'] +
                           tabela_geral['His Qt_acertos'] +
                           tabela_geral['Fis Qt_acertos'] +
                           tabela_geral['Qui Qt_acertos'])
tabela_geral
````

````python
# 3) Seu código aqui
tabela_porcentagem['Total']= (tabela_porcentagem['Mat'] +
                                tabela_porcentagem['Por'] +
                                tabela_porcentagem['Geo'] +
                                tabela_porcentagem['Ing'] +
                                tabela_porcentagem['His'] +
                                tabela_porcentagem['Fis'] +
                                tabela_porcentagem['Qui'] )

tabela_porcentagem['media'] = tabela_porcentagem['Total']/7
tabela_porcentagem
````

````python
# 4) Seu código aqui
aprovados = tabela_porcentagem['media'] >= 45
quantidade_aprovados = aprovados.sum()
quantidade_aprovados
````
